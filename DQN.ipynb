# Random Agent

```python
from tqdm import trange
import numpy as np

def run_random_agent(env, episodes=10, max_steps=2000):
    
    screens = []
    rewards = []

    for episode in range(10):
        episode_reward = 0
        env.reset()

        for step in trange(max_steps):
            screen = env.render(mode='rgb_array')
            screens.append(screen)

            action = env.action_space.sample()

            _, reward, terminal, _ = env.step(action)
            episode_reward += reward

            if terminal: break

        rewards.append(episode_reward)

    return screens, rewards
```

```python
from tqdm import tqdm
import numpy as np
import cv2

def create_video(screens, filename):
    height, width,_=screens[0].shape
    fourcc = cv2.VideoWriter_fourcc('M','P','4','V')
    video=cv2.VideoWriter(filename,fourcc,60,(width,height))

    for screen in tqdm(screens):
        screen = cv2.cvtColor(screen, cv2.COLOR_RGB2BGR)
        video.write(screen)
    
    cv2.destroyAllWindows()
    video.release()
```

```python
from IPython.display import HTML
from base64 import b64encode

def render_video(filename):

    mp4 = open(filename, 'rb').read()
    data_url = "data:video/mp4;base64," + b64encode(mp4).decode()
    return HTML("""
                <video width=400 controls>
                    <source src="%s" type="video/mp4">
                </video>
                """ % data_url
                )
```

```python
def run_custom_environment(env_name):

    env = gym.make(env_name)
    screens, rewards = run_random_agent(env, episodes=10, max_steps=2000)

    print(
        "\nAverage Reward {} +- {}".format(
            np.mean(rewards),
            np.std(rewards)
        )
    )

    create_video(screens, "{}.mp4".format(env_name))

    return render_video("{}.mp4".format(env_name))
```

# DQN

```python
from train import get_train_args
from tt import get_test_args
from play import get_play_args

train_args = get_train_args(['--env', 'Pong-v0'])
test_args = get_test_args(train_args, ['--env', 'Pong-v0'])
play_args = get_play_args(train_args, ['--env', 'Pong-v0'])
```

```python
import tensorflow as tf
from utils.network import DeepQNetwork
import gym

def create_dqn(env, args):

    num_actions = env.action_space.n

    state_ph = tf.placeholder(tf.uint8, (None, play_args.frame_height,
                                         play_args.frame_width,
                                         play_args.frames_per_state))
    
    DQN = DeepQNetwork(num_actions, state_ph, scope='DQN_main')
    return DQN, state_ph
```

```python
def load_session(args):

    config = tf.ConfigProto(allow_soft_placement=True)
    config.gpu_options.allow_growth = True
    sess = tf.Session(config=config)

    loader = tf.train.Saver()
    if args.ckpt_file is not None:
        ckpt = args.ckpt_dir + '/' + args.ckpt_file
    else:
        ckpt = tf.train.latest_checkpoint(args.ckpt_dir)
    
    loader.restore(sess, ckpt)
    print('%s restored.\n\n' % ckpt)

    return sess
```

```python
from utils.state_buffer import StateBuffer
from utils.utils import preprocess_image, reset_env_and_state_buffer

def run_trained_agent(env, DQN, state_ph, sess, args):

    DQN_predict_op = DQN.predict()

    state_buf = StateBuffer(args)
    screens = []
    rewards = []

    for ep in range(0, args.num_eps):
        reset_env_and_state_buffer(env, state_buf, args)
        ep_done = False
        initial_steps = np.random.randint(1, args.max_initial_random_steps+1)
        reward = 0
        for step in tqdm(range(args.max_ep_length)):
            screen = env.render(mode='rgb_array')
            screens.append(screen)

            if step < initial_steps:
                action = env.action_space.sample()
            else:
                state = np.expand_dims(state_buf.get_state(), 0)
                action = sess.run(DQN_predict_op, {state_ph:state})

            
            frame, r, ep_terminal, _ = env.step(action)
            frame = preprocess_image(frame, play_args.frame_width, play_args.frame_height)
            state_buf.add(frame)
            reward += r

            if ep_terminal:
                break
        
        rewards.append(reward)

    return screens, rewards
```

```python
from tt import test

def evaluate_model(args):

    tf.reset_default_graph()
    test(args)
```

```python
from train import train
import tensorflow as tf

def get_custom_args(env_name):

    train_args = get_train_args(['--env', env_name,
                                 '--num_steps_train', '500000',
                                 '--epsilon_step_end', '250000',
                                 '--save_ckpt_step', '10000',
                                 '--save_log_step', '10000',
                                 '--ckpt_dir', './ckpts-{}'.format(env_name),
                                 '--log_dir', './logs-{}/train'.format(env_name)
                            ])
    play_args = get_play_args(train_args, ['--env', env_name,
                                           '--ckpt_dir', './ckpts-{}'.format(env_name)])
    test_args = get_test_args(train_args, ['--env', env_name,
                                           '--ckpt_dir', './ckpts-{}'.format(env_name),
                                            '--log_dir', './log-{}/test'.format(env_name)])
    
    return train_args, play_args, test_args
```

# Custom Environment

```python
def train_custom_environment(env_name):

    train_args, _, _ = get_custom_args(env_name)
    tf.reset_default_graph()
    train(train_args)
```

```python
train_custom_environment('Qbert-v0')
```

```python
def test_custom_environment(env_name):

    _, play_args, test_args = get_custom_args(env_name)

    env = gym.make(env_name)

    tf.reset_default_graph()
    DQN, state_ph = create_dqn(env, play_args)

    sess = load_session(play_args)

    screens, rewards = run_trained_agent(env, DQN, state_ph, sess, play_args)
    print("\nAverage Reward {} +- {}".format(np.mean(rewards), np.std(rewards)))
    create_video(screens, "trained-{}.mp4".format(env_name))

    evaluate_model(test_args)

    return render_video("trained-{}.mp4".format(env_name))
```

```python
test_custom_environment('Qbert-v0')
```
